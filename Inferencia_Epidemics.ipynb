```python
# Caminho raiz
diretorio_raiz = "/home/users/rluna/epidemics/dataset"

# Dicionário para armazenar os modelos
modelos_dict = {}

# Expressão regular para extrair grafo, obs
padrao_nome = re.compile(r"(?P<grafo>.+?)_(TRAIN|TEST)_obs_(?P<obs>\d+)")

for raiz, _, arquivos in os.walk(diretorio_raiz):
    for nome_arquivo in arquivos:
        if nome_arquivo.endswith(".gnn"):
            caminho_completo = os.path.join(raiz, nome_arquivo)

            # Extrai grafo e obs do nome do arquivo
            match = padrao_nome.search(nome_arquivo)
            if not match:
                print(f"Ignorado: {nome_arquivo} (padrão não reconhecido)")
                continue

            grafo = match.group("grafo")
            obs = int(match.group("obs"))

            # Inferência do tipo com base na pasta ou nome do arquivo
            tipo = "treino" if "train" in caminho_completo.lower() else "teste"

            # Carrega o modelo
            try:
                modelo = torch.load(caminho_completo, map_location="cpu")  # ou "cuda"
                chave = nome_arquivo  # ou use (grafo, tipo, obs) como chave, se preferir
                modelos_dict[chave] = {
                    "grafo": grafo,
                    "tipo": tipo,
                    "obs": obs,
                    "modelo": modelo,
                }
            except Exception as e:
                print(f"Erro ao carregar {caminho_completo}: {e}")

print(f"{len(modelos_dict)} modelos carregados.")
```

```python
"""
    File containing the data formats for the project
"""

import os
import pickle
import torch
from torch_geometric.data import Dataset


# TODO: Add logger

class EpidemicInstance:
    def __init__(self, graph, input_metrics, target, metadados):
        self.G = graph
        self.X_not_normalized = input_metrics  # Tensor [n_nodes, n_features]
        self.y = target                         # Ground truth dos infectados
        self.metadados = metadados

        # Copia a coluna 0 sem normalizar
        X0 = self.X_not_normalized[:, 0:1]  # Infectados observáveis (mantido)
        X_rest = self.X_not_normalized[:, 1:]  # Métricas estruturais (normalizar)

        # Normalização z-score nas demais colunas
        mean = X_rest.mean(dim=0)
        std = X_rest.std(dim=0)
        std[std == 0] = 1.0  # Evita divisão por zero

        X_rest_normalized = (X_rest - mean) / std

        # Concatena: coluna 0 original + demais normalizadas
        self.X = torch.cat([X0, X_rest_normalized], dim=1)


class EpidemicDataset(Dataset):
    """
    Dataset para carregar arquivos .pkl contendo instâncias da simulação epidêmica,
    gerando grafos com features para treinar GNNs.
    """

    def __init__(self, folder, inputs):
        """
        Args:
            folder (str): Caminho da pasta contendo os arquivos .pkl.
            inputs (list[str]): Lista com os nomes das features a serem usadas como entrada (ex: ['OBS_I', 'DEG', ...])
        """
        self.path = folder
        self.data = []
        self.inputs = inputs
        if "OBS_I" not in inputs:
            self.inputs = ["OBS_I"] + inputs
            print("⚠️ Warning: observed infected not in the inputs. They will be inserted by default. "
                  f"New input list: {self.inputs}")
        elif inputs[0] != "OBS_I":
            self.inputs.remove("OBS_I")
            self.inputs = ["OBS_I"] + inputs
            print(f"Debug: Reordered input list: {self.inputs}")

        self.obs_b_pos = None if not ("OBS_B" in self.inputs) else self.inputs.index("OBS_B")
        self.cont_pos = None if not ("CONT" in self.inputs) else self.inputs.index("CONT")
        self.cont_k2_pos = None if not ("CONT_k2" in self.inputs) else self.inputs.index("CONT_k2")

        self.read_instances(folder)

    def read_instances(self, folder):
        """
        Lê todos os arquivos .pkl no diretório e extrai os grafos (data_point) com
        features combinadas.
        """
        if not os.path.exists(folder):
            raise ValueError(f"Directory '{folder}' not found")

        for filename in os.listdir(folder):
            if not filename.endswith(".pkl"):
                continue

            path = os.path.join(folder, filename)
            with open(path, "rb") as f:
                ins = pickle.load(f)

            if not isinstance(ins, EpidemicInstance):
                print(f"⚠️ Warning: File '{filename}' is not a valid EpidemicInstance (got {type(ins)}). Skipping.")
                continue

            data_point = ins.G

            # Concatena os tensores de entrada
            try:
                metrics = [ins.X[inp].unsqueeze(-1) for inp in self.inputs]

                X_not_normalized = torch.cat(metrics, dim=-1)  # Tensor [n_nodes, n_features]
                # Copia a coluna 0 sem normalizar
                X0 = X_not_normalized[:, 0:1]  # Infectados observáveis (mantido)
                X_rest = X_not_normalized[:, 1:]  # Métricas estruturais (normalizar)
                # Normalização z-score nas demais colunas
                mean = X_rest.mean(dim=0)
                std = X_rest.std(dim=0)
                std[std == 0] = 1.0  # Evita divisão por zero
        
                X_rest_normalized = (X_rest - mean) / std
                # Concatena: coluna 0 original + demais normalizadas
                X = torch.cat([X0, X_rest_normalized], dim=1)
                
                data_point.x = X#torch.cat(metrics, dim=-1)
                
                data_point.y = ins.y.float().unsqueeze(-1)  # Saída esperada: [N, 1]
                
                if not self.obs_b_pos:
                    data_point.obs_b = ins.X["OBS_B"].unsqueeze(-1)
                if not self.cont_pos:
                    data_point.cont = ins.X["CONT"].unsqueeze(-1)
                if not self.cont_k2_pos:
                    data_point.cont_k2 = ins.X["CONT_k2"].unsqueeze(-1)

                self.data.append(data_point)

            except KeyError as e:
                print(f"❌ KeyError: '{e}' não encontrado em ins.X ao processar '{filename}'. Pulando arquivo.")

    def len(self):
        return len(self.data)
        
    def __len__(self):
        return len(self.data)

    def __getitem__(self, index):
        return self.data[index]

    def all_data(self):
        return self.data
    
    def get(self, idx):
        return self.data[idx]

    def get_observed_betweenness(self, idx):
        if self.obs_b_pos:
            return self.data[idx].x[self.obs_b_pos]
        else:
            return self.data[idx].obs_b

    def get_contact(self, idx):
        if self.cont_pos:
            return self.data[idx].x[self.cont_pos]
        else:
            return self.data[idx].cont

    def get_2neighborhood_contact(self, idx):
        if self.cont_k2_pos:
            return self.data[idx].x[self.cont_k2_pos]
        else:
            return self.data[idx].cont_k2
```

```python
import os
import numpy as np
import pandas as pd
import torch
from sklearn.metrics import roc_auc_score, f1_score, precision_score
from torch_geometric.loader import DataLoader

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

def top_hits_percent(y_true, y_pred, percent=0.01):
    N = len(y_true)
    K = max(1, int(np.floor(N * percent)))
    indices_top_k = np.argsort(y_pred)[::-1][:K]
    hits = np.sum(y_true[indices_top_k])
    return hits / K

def precision_at_percent(y_true, y_pred, percent=0.01):
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    N = len(y_true)
    K = max(1, int(np.floor(N * percent)))
    top_k_indices = np.argsort(y_pred)[::-1][:K]
    true_positives_in_topk = np.sum(y_true[top_k_indices])
    total_positivos = np.sum(y_true)
    if total_positivos == 0:
        return float("nan")
    return true_positives_in_topk / total_positivos

def avaliar_modelo_em_dataset(modelo, dataset):
    modelo.eval()
    y_true_all = []
    y_pred_all = []

    loader = DataLoader(dataset, batch_size=1, shuffle=False)

    with torch.no_grad():
        for batch in loader:
            batch = batch.to(device)
            out = modelo(batch)
            out = out.view(-1).cpu().numpy()
            y = batch.y.view(-1).cpu().numpy()
            x = batch.x[:, 0].cpu().numpy()
        
            mask = x != 1    
            
            y_pred_all.extend(out[mask])
            y_true_all.extend(y[mask])
            

    y_true_all = np.array(y_true_all)
    y_pred_all = np.array(y_pred_all)

    top1 = top_hits_percent(y_true_all, y_pred_all, percent=0.01)
    top5 = top_hits_percent(y_true_all, y_pred_all, percent=0.05)
    top10 = top_hits_percent(y_true_all, y_pred_all, percent=0.10)
    top15 = top_hits_percent(y_true_all, y_pred_all, percent=0.15)
    top20 = top_hits_percent(y_true_all, y_pred_all, percent=0.20)

    try:
        auc = roc_auc_score(y_true_all, y_pred_all)
    except:
        auc = float("nan")
    try:
        f1 = f1_score(y_true_all, y_pred_all > 0.5)
        precision = precision_score(y_true_all, y_pred_all > 0.5)
    except:
        f1 = precision = float("nan")

    prec_at_1 = precision_at_percent(y_true_all, y_pred_all, percent=0.01)
    prec_at_5 = precision_at_percent(y_true_all, y_pred_all, percent=0.05)
    prec_at_10 = precision_at_percent(y_true_all, y_pred_all, percent=0.10)
    prec_at_15 = precision_at_percent(y_true_all, y_pred_all, percent=0.15)
    prec_at_20 = precision_at_percent(y_true_all, y_pred_all, percent=0.20)
    return {
        "top1": top1,
        "top5": top5,
        "top10": top10,
        "top15": top15,
        "top20": top20,
        "auc": auc,
        "f1": f1,
        "precision": precision,
        "precision@1%": prec_at_1,
        "precision@5%": prec_at_5,
        "precision@10%": prec_at_10,
        "precision@15%": prec_at_15,
        "precision@20%": prec_at_20,
    }

# Base das subpastas
base_path = "/home/users/rluna/epidemics/dataset/Testes/BA"

# Lista para armazenar os resultados
linhas_resultado = []

for chave, info in modelos_dict.items():
    modelo = info["modelo"].to(device)
    obs = info["obs"]  # ex: 75

    for subpasta in os.listdir(base_path):
        if not subpasta.startswith("teste_"):
            continue

        caminho = os.path.join(base_path, subpasta)
        print(f"\n Modelo: {chave} → Subpasta: {subpasta}")

        try:
            dataset = EpidemicDataset(caminho,input_fields)
            if len(dataset) == 0:
                print(f" Subpasta vazia: {caminho}")
                continue

            metricas = avaliar_modelo_em_dataset(modelo, dataset)
            metricas["modelo"] = chave
            metricas["subpasta"] = subpasta
            linhas_resultado.append(metricas)

            print(f" Métricas: {metricas}")

        except Exception as e:
            print(f" Erro ao processar {caminho}: {e}")

# Criar DataFrame e salvar
df = pd.DataFrame(linhas_resultado)
df.to_csv("avaliacoes_modelos_por_subpasta_BA.csv", index=False)
print(" CSV salvo: avaliacoes_modelos_por_subpasta.csv")
```
